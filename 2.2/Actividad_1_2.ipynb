```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
```

```python
#Carga desde un archivo .csv sin indice
df = pd.read_csv("inmuebles2.csv")
df
```

```python
#a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” de “Referencia”
df1 = pd.read_csv("inmuebles2.csv", skiprows=1)
df1
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
df2 = df1.drop([df1.columns[0],'Referencia'], axis=1)
df2
```

```python
#2 Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00
filtro2 = df2[(df2['Fecha Alta'] == "2007-04-29 00:00:00") & (df2['Fecha Alta'] == "2004-01-02 00:00:00")]
filtro2
```

```python
# 3 Todos los registros de “ Tipo: parking, local, industrial y piso”
filtro3 = df2[(df2["Tipo"] == "Parking") | (df2["Tipo"] == "Local") | (df2["Tipo"] == "Industrial") | (df2["Tipo"] == "Piso")]
filtro3
```

```python
# 4 Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”
filtro4 = df2[(df2['Provincia'] == "Girona") & (df2['Superficie'] > 100)]
print(filtro4)
```

```python
# 5 Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”
filtro5 = df2[(df2["Precio Venta"] > 1000000) | (df2["Vendedor"].isin(["Carmen", "Joaquin"]))]
print(filtro5)
```

```python
# 6 Todos los registros que no sean de “Provincia : Barcelona“
filtro6 = df2[~(df2["Provincia"] == "Barcelona")]
print(filtro6)
```

```python
#7 Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“
filtro7 = df2[~(df2["Operación"] != "Alquiler") & ~(df2["Tipo"] != "Oficina")]
print(filtro7)
```

```python
filtro2.to_csv("RetoFiltro2.2.csv")
filtro3.to_csv("RetoFiltro2.3.csv")
filtro4.to_csv("RetoFiltro2.4.csv")
filtro5.to_csv("RetoFiltro2.5.csv")
filtro6.to_csv("RetoFiltro2.6.csv")
filtro7.to_csv("RetoFiltro2.7.csv")
```
